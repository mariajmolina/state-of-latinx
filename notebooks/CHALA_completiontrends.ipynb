```python
import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
import completions_module as cmodule
```

```python
all_files = glob.glob("../data/completions_*.csv")
all_files = sorted(all_files)
```

```python
# ---

df = cmodule.open_and_concat(all_files)
df = cmodule.all_degreesandcerts(df)
df = cmodule.physical_sciences(df)

# ---

df_nonhis = cmodule.diff(
                 df.groupby('year').sum()['Grand total'], 
                 df.groupby('year').sum()['Hispanic or Latino total'], 
                 'Non-Hispanic total')['Non-Hispanic total']

df_nonhismen = cmodule.diff(
                    df.groupby('year').sum()['Grand total men'], 
                    df.groupby('year').sum()['Hispanic or Latino men'], 
                    'Non-Hispanic men')['Non-Hispanic men']

df_nonhiswomen = cmodule.diff(
                      df.groupby('year').sum()['Grand total women'], 
                      df.groupby('year').sum()['Hispanic or Latino women'], 
                      'Non-Hispanic women')['Non-Hispanic women']

# ---

df_his = df.groupby('year').sum()['Hispanic or Latino total']

df_hismen = df.groupby('year').sum()['Hispanic or Latino men']

df_hiswomen = df.groupby('year').sum()['Hispanic or Latino women']

# ---

df_his.plot(figsize=(10,6), subplots=False, label='Hispanic/Latinx',
            color=['darkgrey','slateblue','indigo','thistle','k'])

plt.legend(loc='upper left', ncol=3)
plt.xlabel('Years', fontsize=14)
plt.xticks([2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019], 
           [2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019], 
           fontsize=12)

plt.yticks(fontsize=12)
plt.ylim([0,2200])
plt.title('All Degree and Certificate Completions in Physical Sciences (2009-2019)', fontsize=14)
plt.grid('0.15', ls='--', lw=0.5)
#plt.savefig('../figures/bars_chala.png', bbox_inches='tight', dpi=200)
plt.show()

# ----

df_hiswomen.plot(figsize=(10,6), subplots=False, label='Hispanic/Latinx Women', color=['k'])

df_hismen.plot(figsize=(10,6), subplots=False, label='Hispanic/Latinx Men', color=['slateblue'])

plt.legend(loc='upper left', ncol=3)
plt.xlabel('Years', fontsize=14)
plt.xticks([2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019], 
           [2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019], 
           fontsize=12)

plt.yticks(fontsize=12)
plt.ylim([0,1500])
plt.title('All Degree and Certificate Completions in Physical Sciences (2009-2019)', fontsize=14)
plt.grid('0.15', ls='--', lw=0.5)
#plt.savefig('../figures/bars_chala.png', bbox_inches='tight', dpi=200)
plt.show()

# ----
```
